```python
import torch
import matplotlib.pyplot as plt
```

```python
efficient_style = torch.load("/content/drive/MyDrive/efficient_style.pth")

train_losses = efficient_style['train_losses']
train_accuracies = efficient_style['train_accuracies']
val_losses = efficient_style['val_losses']
val_accuracies = efficient_style['val_accuracies']

# Plot Accuracies
plt.figure(figsize=(8,5))
plt.title("Style")
plt.plot(train_accuracies,label="Train Accuracy")
plt.plot(val_accuracies,label="Val Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
