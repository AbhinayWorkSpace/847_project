```python
import torch
import matplotlib.pyplot as plt
```

```python
vit_style = torch.load("/content/drive/MyDrive/vit_style.pth")

train_losses = vit_style['train_losses']
train_accuracies = vit_style['train_accuracies']
val_losses = vit_style['val_losses']
val_accuracies = vit_style['val_accuracies']

# Plot Accuracies
plt.figure(figsize=(8,5))
plt.title("Style")
plt.plot(train_accuracies,label="Train Accuracy")
plt.plot(val_accuracies,label="Val Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
